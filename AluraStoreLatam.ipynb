### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

#1. Análisis de facturación



```python
# 1. Definimos las tiendas y sus nombres
lista_dfs = [tienda, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# 2. Calculamos la suma de la columna 'Precio' para cada una
ventas_totales = [df['Precio'].sum() for df in lista_dfs]

# 3. Creamos una tablita resumen
df_facturacion = pd.DataFrame({
    'Tienda': nombres,
    'Total_Ventas': ventas_totales
})

df_facturacion
```

```python
df_facturacion = df_facturacion.sort_values(by='Total_Ventas', ascending=False)
df_facturacion
```

```python
df_facturacion.plot(x='Tienda',y='Total_Ventas',kind='bar',figsize=(8, 3),title='Facturación Total por Tienda ',
                    ylabel='Ventas ($)',xlabel= 'Tienda',color='blue',rot=0)
```

###**Interpretación de las ventas totales**

Al analizar el volumen de ingresos por sucursal, se observa un patrón lineal descendente: la Tienda 1 lidera las ventas, seguida de manera sucesiva por la Tienda 2 y la Tienda 3, dejando a la Tienda 4 con el desempeño más bajo de todo el grupo.

# 4. Productos más y menos vendidos

```python
# 1. Calculamos la frecuencia de cada producto por tienda
frec_t1 = tienda['Producto'].value_counts()
frec_t2 = tienda2['Producto'].value_counts()
frec_t3 = tienda3['Producto'].value_counts()
frec_t4 = tienda4['Producto'].value_counts()

# 2. Creamos la tabla de los MÁS VENDIDOS (Top 3)
df_mas_vendidos = pd.DataFrame({
    'Tienda 1': frec_t1.head(3).index,
    'Tienda 2': frec_t2.head(3).index,
    'Tienda 3': frec_t3.head(3).index,
    'Tienda 4': frec_t4.head(3).index})

# 3. Creamos la tabla de los MENOS VENDIDOS (Bottom 3)
df_menos_vendidos = pd.DataFrame({
    'Tienda 1': frec_t1.tail(3).index,
    'Tienda 2': frec_t2.tail(3).index,
    'Tienda 3': frec_t3.tail(3).index,
    'Tienda 4': frec_t4.tail(3).index
})


display(df_mas_vendidos)

display(df_menos_vendidos)
```

###**Interpretación de los Datos de ventas por producto más y menos vendidos**

Al analizar específicamente los nombres de los productos, observamos una dinámica de inventario interesante:

Alta Rotación (Top 3): Los productos más vendidos coinciden con las categorías de Electrónica y Muebles, destacando artículos como microondas, mesa de comedor, etc. Esto confirma que el flujo de clientes está dirigido a bienes durables.

Baja Rotación: Identificamos productos que apenas han tenido ventas en el periodo, principalmente en las categorías de Libros y Juguetes.

Conclusión Estratégica: La Tienda 4 presenta productos en su 'Top 3' que son similares a las tiendas más exitosas, lo que refuerza la idea de que su problema no es la selección de productos (el 'qué' vende), sino el volumen de clientes que entran a la tienda (el 'cuánto' vende).

# 5. Envío promedio por tienda

```python
# 1. Calculamos el promedio de la columna 'Valor del envío' para cada tienda
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

# 2. Creamos el DataFrame para comparar
df_envios = pd.DataFrame({'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
                          'Envío Promedio ($)': [envio_t1, envio_t2, envio_t3, envio_t4]})

# 3. Mostramos la tabla formateada
display(df_envios)

# 4. Generamos el gráfico de barras
df_envios.plot(x='Tienda', y='Envío Promedio ($)', kind='bar', figsize=(8, 5),color='orange',
               title='Comparativa de Costos de Envío',ylabel='Costo Promedio ($)',rot=0)
```

```python
# 1. Calculamos la Ganancia Neta para cada registro de cada tienda
# (Asumiendo que el costo de envío está dentro del precio de venta)
tienda['Ganancia_Neta'] = tienda['Precio'] - tienda['Costo de envío']
tienda2['Ganancia_Neta'] = tienda2['Precio'] - tienda2['Costo de envío']
tienda3['Ganancia_Neta'] = tienda3['Precio'] - tienda3['Costo de envío']
tienda4['Ganancia_Neta'] = tienda4['Precio'] - tienda4['Costo de envío']

# 2. Sumamos las ganancias para comparar el total "real"
ganancia_t1 = tienda['Ganancia_Neta'].sum()
ganancia_t2 = tienda2['Ganancia_Neta'].sum()
ganancia_t3 = tienda3['Ganancia_Neta'].sum()
ganancia_t4 = tienda4['Ganancia_Neta'].sum()

# 3. Creamos un DataFrame comparativo
df_margen = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ventas Brutas': [tienda['Precio'].sum(), tienda2['Precio'].sum(), tienda3['Precio'].sum(), tienda4['Precio'].sum()],
    'Ganancia Neta Total': [ganancia_t1, ganancia_t2, ganancia_t3, ganancia_t4]})

# Calculamos el % de eficiencia (Qué tanto de la venta es ganancia real)
df_margen['Eficiencia (%)'] = (df_margen['Ganancia Neta Total'] / df_margen['Ventas Brutas']) * 100

display(df_margen)

# 4. Gráfico de barras comparativo
df_margen.plot(x='Tienda', y=['Ventas Brutas', 'Ganancia Neta Total'], kind='bar', figsize=(10,6),
               title='Diferencia entre Facturación y Ganancia Real', color=['#007bff', '#28a745'])
```

###**Interpretación de los Datos de envío promedio por tienda**

El impacto del gasto logístico en cada sucursal es fundamental. Dado que el cliente es quien asume el costo del envío, este valor funciona como un indicador de competitividad geográfica:

Liderazgo en Costos: La Tienda 4 registra el costo de envío promedio más bajo de toda la red, comparado con los de la Tienda 1.

Eficiencia de Margen: Al realizar el ejercicio de restar el costo de envío al valor de venta, observamos que todas las tiendas mantienen una eficiencia cercana al 94.6%. Esto significa que la Tienda 4 no está perdiendo dinero por cada venta; de hecho, es la que ofrece el servicio de entrega más económico para el bolsillo del cliente.

Barrera de Entrada: El hecho de que la Tienda 4 sea la más barata en envíos descarta la logística como el problema. Si el cliente paga menos por el despacho en esta tienda que en las demás, el bajo volumen de ventas debe atribuirse a factores externos como la densidad poblacional o la falta de visibilidad del local.

Conclusión del Ítem: La Tienda 4 es operativamente la más eficiente en términos de despacho. Su bajo rendimiento no es un problema de costos elevados, sino de falta de masa crítica de pedidos.

#6. **Análisis del desempeño geográfico**

```python
import matplotlib.pyplot as plt

# Creamos una figura con 4 sub-gráficos
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)
axes = axes.flatten()
tiendas_dfs = [tienda, tienda2, tienda3, tienda4]
colores = ['blue', 'orange', 'green', 'red']

for i, df in enumerate(tiendas_dfs):
    axes[i].scatter(df['lon'], df['lat'], color=colores[i], alpha=0.4, s=20)
    axes[i].set_title(f'Ventas Geográficas: Tienda {i+1}')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

###**Interpretación de la Distribución Geográfica de las Ventas**

Al mapear las coordenadas de latitud y longitud de cada pedido, obtenemos una visión clara de la cobertura territorial de nuestras sucursales:

Solapamiento de Zonas: Los gráficos muestran que las cuatro tiendas compiten prácticamente por el mismo territorio. No hay una "zona exclusiva" para la Tienda 4; sus entregas ocurren en las mismas latitudes y longitudes que las tiendas más grandes.

Densidad de Pedidos: Al comparar los mapas individuales, se observa que la Tienda 1 tiene una concentración de puntos mucho más densa (más clientes por kilómetro cuadrado). En cambio, la Tienda 4 tiene puntos más dispersos.

Veredicto Geográfico: El bajo rendimiento de la Tienda 4 no se debe a que esté en un lugar "malo" o aislado, sino a que está subutilizada dentro de una zona donde otras sucursales ya tienen presencia consolidada. Los clientes que captan son fieles y están satisfechos, pero son menos en cantidad.

Conclusión del Ítem: La ubicación actual de la Tienda 4 la obliga a competir directamente con el volumen de la Tienda 1. La baja venta es un problema de cuota de mercado, no de eficiencia operativa ni de falta de demanda en la zona.
